# **Data Cleaning**

```python
mrgdf1 = mrgdf.drop("district" , axis = "columns")
```

```python
mrgdf1
```

```python
mrgdf1["review_scores_rating"] = mrgdf1["review_scores_rating"].fillna(mrgdf1["review_scores_rating"].mean())
mrgdf1["review_scores_accuracy"] = mrgdf1["review_scores_accuracy"].fillna(mrgdf1["review_scores_accuracy"].mean())
mrgdf1["review_scores_cleanliness"] = mrgdf1["review_scores_cleanliness"].fillna(mrgdf1["review_scores_cleanliness"].mean())
mrgdf1["review_scores_checkin"] = mrgdf1["review_scores_checkin"].fillna(mrgdf1["review_scores_checkin"].mean())
mrgdf1["review_scores_communication"] = mrgdf1["review_scores_communication"].fillna(mrgdf1["review_scores_communication"].mean())
mrgdf1["review_scores_location"] = mrgdf1["review_scores_location"].fillna(mrgdf1["review_scores_location"].mean())
mrgdf1["review_scores_value"] = mrgdf1["review_scores_value"].fillna(mrgdf1["review_scores_value"].mean())
```

```python
mrgdf1.isnull().sum()
```

```python
mrgdf1["bedrooms"]
```

```python
mrgdf1["bedrooms"] = mrgdf1["bedrooms"].fillna(mrgdf1["bedrooms"].mean())
```

```python
mrgdf1.isnull().sum()
```

```python
mrgdf1["host_identity_verified"] = mrgdf1["host_identity_verified"].fillna(mrgdf1["host_identity_verified"].mode()[0])
mrgdf1["host_response_time"] = mrgdf1["host_response_time"].fillna(mrgdf1["host_response_time"].mode()[0])
mrgdf1["host_location"] = mrgdf1["host_location"].fillna(mrgdf1["host_location"].mode()[0])
mrgdf1["host_is_superhost"] = mrgdf1["host_is_superhost"].fillna(mrgdf1["host_is_superhost"].mode()[0])
```

```python
mrgdf1.isnull().sum()
```

```python
mrgdf1 = mrgdf1.replace({"t": 1 , "f" : 0})
```

```python
mrgdf1[["instant_bookable" , "host_identity_verified"]]
```

```python
mrgdf1["host_response_rate"] = mrgdf1["host_response_rate"].fillna(mrgdf1["host_response_rate"].mean())
mrgdf1["host_acceptance_rate"] = mrgdf1["host_acceptance_rate"].fillna(mrgdf1["host_acceptance_rate"].mean())
mrgdf1["host_total_listings_count"] = mrgdf1["host_total_listings_count"].fillna(mrgdf1["host_total_listings_count"].mean())
mrgdf1["host_has_profile_pic"] = mrgdf1["host_has_profile_pic"].fillna(mrgdf1["host_has_profile_pic"].mean())
```

```python
mrgdf1.isnull().sum()
```

```python
mrgdf1["host_is_superhost"]
```

```python
mrgdf1["name"]
```

```python
mrgdf1["name"] = mrgdf1["name"].fillna(mrgdf1["name"].mode()[0])
```

```python
mrgdf1["host_since"] = pd.to_datetime(mrgdf1["host_since"])


avg_date = mrgdf1['host_since'].mean()
mrgdf1["host_since"].fillna(avg_date, inplace=True)
```
